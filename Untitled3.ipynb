```python
# Install required libraries
!pip install datasets scipy --quiet

# Import necessary modules
from datasets import load_dataset
import numpy as np
import pandas as pd
from scipy.signal import butter, lfilter
import random
import matplotlib.pyplot as plt

# Helper functions
length = 1000  # e.g., 10 seconds at 100Hz
t = np.linspace(0, 10, length)
waveform = np.sin(2 * np.pi * 5 * t) * np.exp(-0.05 * t) + np.random.normal(0, 0.05, length)

def low_pass_filter(data, cutoff=0.05, fs=100.0, order=4):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return lfilter(b, a, data)

def add_damage_impulse(data, magnitude_range=(1.0, 3.0), duration_range=(10, 50)):
    data = data.copy()
    length = len(data)
    # Avoid damage injection near edges
    start = random.randint(100, length - 100)
    duration = random.randint(*duration_range)
    magnitude = random.uniform(*magnitude_range)
    end = min(start + duration, length)
    data[start:end] += magnitude
    return data

# Load dataset directly from Huggingface hub
dataset = pd.read_csv("/content/query_M2.5%2B_2000-2024.csv")

print(f"Dataset loaded with {len(dataset)} samples")

# Process first 10 samples for demo
all_records = []

for idx, sample in dataset.iloc[:10].iterrows():
    sensor_healthy = waveform
    sensor_damaged = add_damage_impulse(sensor_healthy)
    sensor_damped = low_pass_filter(sensor_healthy)

    time = np.arange(length) / 100.0  # Assuming 100 Hz sampling rate

    df_sample = pd.DataFrame({
        'time': time,
        'sensor_healthy': sensor_healthy,
        'sensor_damaged': sensor_damaged,
        'sensor_damped': sensor_damped,
        'sample_id': idx
    })

    # Convert to long format for convenience
    df_long = df_sample.melt(id_vars=['time', 'sample_id'],
                             value_vars=['sensor_healthy', 'sensor_damaged', 'sensor_damped'],
                             var_name='sensor_type', value_name='amplitude')

    df_long['label'] = df_long['sensor_type'].apply(lambda x: x.split('_')[1])
    all_records.append(df_long)

# Combine all samples
df_all = pd.concat(all_records, ignore_index=True)

# Save to CSV
df_all.to_csv('synthetic_shm_dataset.csv', index=False)

print("Synthetic SHM dataset created and saved as 'synthetic_shm_dataset.csv'")

# Plot a sample for visualization
sample_plot = df_all[df_all['sample_id'] == 0]

plt.figure(figsize=(12,6))
for label in sample_plot['label'].unique():
    subset = sample_plot[sample_plot['label'] == label]
    plt.plot(subset['time'], subset['amplitude'], label=label)
plt.title('Synthetic SHM Sensor Signals from SeismicDataWorldwide Sample 0')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
